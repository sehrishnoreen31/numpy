```python
import numpy as np
import sympy as sp
from sympy import Matrix, pprint
import pandas as pd
```

```python
A = np.array([[3, 4, 2], [3, 2, 0], [5, 6, 7]])
A
```

```python
# as dataframe
df = pd.DataFrame(A)
display(df)
```

## Row Echelon


```python
row_ech_form_A = Matrix(A).echelon_form()
display(row_ech_form_A)
```

```python
row_ech_form_A = Matrix(df).echelon_form()
display(row_ech_form_A)
```

## Reduced Row Echelon Form


```python
reduced_row_ech_form = Matrix(A).rref()
print(
    "Reduced Row Echelon Form:",
)
# pprint(reduced_row_ech_form[0])
# reduced_row_ech_form[0]
display(reduced_row_ech_form[0])
print("Solution:")
display(reduced_row_ech_form[1])
```

```python
# as df
reduced_row_ech_form = Matrix(df).rref()
print(
    "Reduced Row Echelon Form:",
)
# pprint(reduced_row_ech_form[0])
# reduced_row_ech_form[0]
display(reduced_row_ech_form[0])
print("Solution:")
display(reduced_row_ech_form[1])
```
