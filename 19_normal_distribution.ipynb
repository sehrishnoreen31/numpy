```python
import numpy as np
import seaborn as sns
```

```python
x = np.random.normal(size=(4,4))
display(x)
sns.displot(x, kind='kde')
```

```python
my_array = np.random.normal(size=(4,4), scale=2)
sns.displot(my_array, kind='kde')
```

```python
my_array = np.random.normal(size=(4,4), loc=3, scale=2)
sns.displot(my_array, kind='kde')
```

```python
my_array = np.random.normal(size=(200), scale=2)
sns.displot(my_array, kind='kde')
```
